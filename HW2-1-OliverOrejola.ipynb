Import Relevant Libraries

```python
from tensorflow.keras.datasets import mnist
import matplotlib.pyplot as plt
```

```python
(train_images, train_labels), (test_images, test_labels) = mnist.load_data()
```

Extract the first 10 images of each digit from MNIST training set

```python
index_list= [[],[],[],[],[],[],[],[],[],[]] #Initialize blank list of lists to store indices
for idx, val in enumerate(train_labels):
    if val == 0 and len(index_list[0])<=9:
        index_list[0].append(idx)
    elif val == 1 and len(index_list[1])<=9:
        index_list[1].append(idx)
    elif val == 2 and len(index_list[2])<=9:
        index_list[2].append(idx)
    elif val == 3 and len(index_list[3])<=9:
        index_list[3].append(idx)
    elif val == 4 and len(index_list[4])<=9:
        index_list[4].append(idx)
    elif val == 5 and len(index_list[5])<=9:
        index_list[5].append(idx)
    elif val == 6 and len(index_list[6])<=9:
        index_list[6].append(idx)
    elif val == 7 and len(index_list[7])<=9:
        index_list[7].append(idx)
    elif val == 8 and len(index_list[8])<=9:
        index_list[8].append(idx)    
    elif val == 9 and len(index_list[9])<=9:
        index_list[9].append(idx)
```

```python
fig, axes = plt.subplots(nrows=10, ncols=10)

for row in range(10):
    for col in range(10):
        axes[row,col].imshow(train_images[index_list[col][row]], cmap=plt.cm.binary) 
        axes[row,col].axis('off')
plt.show()
```
